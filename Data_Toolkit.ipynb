# Data Toolkit


1.  What is NumPy, and why is it widely used in Python?
- NumPy is a Python library for numerical computing. It provides high-performance multi-dimensional arrays and functions to perform mathematical operations efficiently.
- Widely used because it is fast, supports vectorized operations, integrates with other libraries (Pandas, Scikit-learn), and is essential for data science.
2.  How does broadcasting work in NumPy?
-   Broadcasting allows NumPy to perform operations on arrays of different shapes without explicitly replicating data.
-    Example:
    a = np.array([1, 2, 3])
    b = 2
    print(a + b)  # [3, 4, 5]

3.  A What is a Pandas DataFrame?
-   A DataFrame is a 2D labeled data structure in Pandas, similar to an Excel sheet or SQL table. It contains rows and columns and supports data manipulation, cleaning, and analysis.
4.  Explain the use of the groupby() method in Pandas.
-  groupby() is used to split data into groups based on a condition, apply functions (like sum, mean, count), and combine results.
-  Example: df.groupby("Category")["Sales"].sum()
5.  Why is Seaborn preferred for statistical visualizations?
-  Seaborn is built on Matplotlib but provides prettier and statistical-friendly plots (heatmaps, violin plots, box plots). It requires fewer lines of code for complex visualizations.
6.  What are the differences between NumPy arrays and Python lists?
-  NumPy arrays: Homogeneous, fixed size, faster, supports mathematical operations.

- Python lists: Heterogeneous, dynamic, slower, don’t support vectorized operations.
7.  What is a heatmap, and when should it be used?
-  A heatmap is a graphical representation of data using color intensity to show values in a matrix.
Used for correlation matrices, feature importance, confusion matrices.
8.  What does the term “vectorized operation” mean in NumPy?
- It means applying operations on entire arrays at once without using explicit loops. This makes computations much faster.
9.  How does Matplotlib differ from Plotly?
- Matplotlib: Static, traditional plots, widely used in Python.

- Plotly: Interactive, web-based visualizations with zoom, hover, and 3D support.
10. What is the significance of hierarchical indexing in Pandas?
-  Allows multiple levels of indexing in rows/columns. Useful for handling multi-dimensional data in a 2D DataFrame.
11. What is the role of Seaborn’s pairplot() function?
-  Creates scatterplots between all numerical variables in a dataset along with histograms on the diagonal → helps visualize relationships.
12. What is the purpose of the describe() function in Pandas?
- Provides summary statistics (mean, median, std, min, max, quartiles) for numerical columns in a DataFrame.
13. Why is handling missing data important in Pandas?
- Missing values can cause incorrect results in calculations and model training. Handling them ensures data accuracy and reliability.
14. What are the benefits of using Plotly for data visualization?
-  Interactive plots

- Supports 3D & dashboards
- Easy integration with Jupyter/Flask/Dash
15. How does NumPy handle multidimensional arrays?
- NumPy uses ndarray objects, supporting multi-dimensional data (2D, 3D, nD) with efficient storage and operations.
16. What is the role of Bokeh in data visualization?
- Bokeh is a Python library for interactive, web-based visualizations. It allows building dashboards easily.
17. Explain the difference between apply() and map() in Pandas.
- map() → works on Series (element-wise).

- apply() → works on Series & DataFrame (row/column-wise functions).
18. What are some advanced features of NumPy?
- Linear algebra (np.linalg)

- Fourier transform (np.fft)

- Random number generation (np.random)

- Broadcasting and masked arrays
19. How does Pandas simplify time series analysis?
- Pandas provides DatetimeIndex, resampling, shifting, and rolling window functions, making time-series analysis simple.
20. What is the role of a pivot table in Pandas?
-  Summarizes data by categories → similar to Excel Pivot Table. Example: df.pivot_table(values="Sales", index="Region", columns="Product")
21. Why is NumPy’s array slicing faster than Python’s list slicing?
- Because NumPy uses contiguous memory blocks (C implementation), while Python lists store pointers → making NumPy much faster.
22. What are some common use cases for Seaborn?
- Correlation heatmaps

- Distribution plots

- Regression plots

- Categorical comparisons (box/violin plots)

# Practical

```python
# 1. How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np
arr = np.array([[1, 2, 3], [4, 5, 6]])
print("Row sums:", arr.sum(axis=1))
```

```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame?
import pandas as pd
df = pd.DataFrame({"Name": ["A", "B", "C"], "Marks": [85, 90, 95]})
print("Mean of Marks:", df["Marks"].mean())
```

```python
# 3. Create a scatter plot using Matplotlib?
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
import numpy as np
import pandas as pd
df = pd.DataFrame(np.random.rand(5,5), columns=list("ABCDE"))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
```

```python
# 5. Generate a bar plot using Plotly.
import plotly.express as px
df = pd.DataFrame({"Name":["A","B","C"], "Score":[80,90,70]})
fig = px.bar(df, x="Name", y="Score", title="Bar Plot")
fig.show()
```

```python
# 6. Create a DataFrame and add a new column based on an existing column.
df["Percentage"] = df["Score"] / df["Score"].sum() * 100
print(df)
```

```python
# 7. Write a program to perform element-wise multiplication of two NumPy arrays?
a = np.array([1,2,3])
b = np.array([4,5,6])
print("Multiplication:", a * b)
```

```python
# 8. Create a line plot with multiple lines using Matplotlib?
x = [1,2,3,4,5]
plt.plot(x, [i*2 for i in x], label="2x")
plt.plot(x, [i*3 for i in x], label="3x")
plt.legend()
plt.show()
```

```python
# 9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
df = pd.DataFrame({"Name":["A","B","C"], "Age":[20,25,30]})
print(df[df["Age"] > 22])
```

```python
# 10. Create a histogram using Seaborn to visualize a distribution?
sns.histplot(df["Age"], bins=5, kde=True)
```

```python
# 11. Perform matrix multiplication using NumPy.
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print(np.dot(A,B))
```

```python
# 12. Use Pandas to load a CSV file and display its first 5 rows.
import pandas as pd

data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Age": [20, 25, 30, 22, 28],
    "Marks": [85, 90, 88, 76, 95]
}

df = pd.DataFrame(data)
df.to_csv("data.csv", index=False)

df2 = pd.read_csv("data.csv")
print(df2.head())

```

```python
# 13. Create a 3D scatter plot using Plotly.
import plotly.express as px
df = pd.DataFrame({"x":[1,2,3,4,5], "y":[2,4,6,8,10], "z":[5,4,3,2,1]})
fig = px.scatter_3d(df, x="x", y="y", z="z", color="z", size="y")
fig.show()
```
